### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
# Inspect injury-related columns
injury_cols = ['total_fatal_injuries', 'total_serious_injuries', 
               'total_minor_injuries', 'total_uninjured']

# Show missing data percentages
print("Missing Values (%):")
missing_pct = (airline_df[injury_cols].isnull().sum() / len(airline_df) * 100)
print(missing_pct)

print("\nBasic Statistics:")
print(airline_df[injury_cols].describe())
```

```python
# Fill missing injury values with 0 (assumption: if not reported, assume 0)
# This is reasonable as fatal/serious injuries would likely be reported
airline_df['total_fatal_injuries'] = airline_df['total_fatal_injuries'].fillna(0)
airline_df['total_serious_injuries'] = airline_df['total_serious_injuries'].fillna(0)
airline_df['total_minor_injuries'] = airline_df['total_minor_injuries'].fillna(0)
airline_df['total_uninjured'] = airline_df['total_uninjured'].fillna(0)

# Calculate total passengers (sum of all injury categories)
airline_df['total_passengers'] = (airline_df['total_fatal_injuries'] + 
                                   airline_df['total_serious_injuries'] + 
                                   airline_df['total_minor_injuries'] + 
                                   airline_df['total_uninjured'])

# Calculate serious/fatal injuries combined
airline_df['fatal_serious_injuries'] = airline_df['total_fatal_injuries'] + airline_df['total_serious_injuries']

# Calculate fraction of passengers with fatal/serious injuries
# Returns NaN automatically where total_passengers = 0 (division by zero)
airline_df['fatal_serious_fraction'] = (
    airline_df['fatal_serious_injuries'] / airline_df['total_passengers']
)

"""New injury metrics created:
  total_passengers: sum of all injury categories
  fatal_serious_injuries: fatal + serious injuries combined
  fatal_serious_fraction: fraction of passengers with fatal/serious injuries"""


print(f"\nSample statistics for fatal_serious_fraction:")
print(airline_df['fatal_serious_fraction'].describe())
```

New injury metrics created:
  total_passengers: sum of all injury categories
  fatal_serious_injuries: fatal + serious injuries combined
  fatal_serious_fraction: fraction of passengers with fatal/serious injuries

**Make Column Cleaning Tasks Identified:**
1. Drop rows with missing Make values

2. Consolidate manufacturer name variations (e.g. "Beech" vs "Beechcraft")

```python
# 1. Drop missing Make values
airline_df = airline_df.dropna(subset=['make'])

# Standardize to title case first (handles ALL CAPS problem)
airline_df['make'] = airline_df['make'].str.strip().str.title()


# 2. Consolidate known manufacturer name variations
make_replacements = {
    'Beech': 'Beechcraft',
    'Mcdonnell Douglas': 'McDonnell Douglas',
    'Mcdonnell Douglas Aircraft Co': 'McDonnell Douglas',
    'Boeing Co': 'Boeing',
    'Piper Aircraft': 'Piper',
    'Grumman American': 'Grumman',
    'Cirrus Design Corp': 'Cirrus',
    'Air Tractor Inc': 'Air Tractor',
    'Bombardier Inc': 'Bombardier',  # add this
    'Aviat Aircraft Inc': 'Aviat'
}
airline_df['make'] = airline_df['make'].replace(make_replacements)

# 5. Keep only makes with 50+ accidents for statistical robustness
make_counts = airline_df['make'].value_counts()
valid_makes = make_counts[make_counts >= 50].index
airline_df = airline_df[airline_df['make'].isin(valid_makes)]

print(f"Remaining records: {len(airline_df)}")
print(f"Remaining unique makes: {airline_df['make'].nunique()}")
print("\nMakes kept:")
print(airline_df['make'].value_counts())
```

###now down to the cleaning required

1. Fix weather_condition - consolidate Unk/UNK into one
2. Fix engine_type - consolidate Unknown/UNK into one
3. Fix number_of_engines - 0 engines is impossible, replace with NaN
4. Fix broad_phase_of_flight - standardize Unknown

```python
# Fix weather_condition - consolidate Unk/UNK into one
airline_df['weather_condition'] = airline_df['weather_condition'].replace({'Unk': 'UNK', 'Unk': 'UNK'})

# Fix engine_type - consolidate Unknown/UNK into one
airline_df['engine_type'] = airline_df['engine_type'].replace({'Unk': 'Unknown'})

# Fix number_of_engines - 0 engines is impossible, replace with NaN
airline_df['number_of_engines'] = airline_df['number_of_engines'].replace(0.0, np.nan)

# Fix broad_phase_of_flight - standardize Unknown
airline_df['broad_phase_of_flight'] = airline_df['broad_phase_of_flight'].replace({'UNKNOWN': 'Unknown'})


for col in cols:
    print(f"\n{col}:")
    print(airline_df[col].value_counts(dropna=False))
```
